```python
# check the null values available in the data
dataset.isnull().sum()
```

```python
# fill the value in Item_Weight column by using mean
dataset['Item_Weight']=dataset['Item_Weight'].fillna(dataset['Item_Weight'].mean())
```

```python
# fill the value in Outlet_Size column by using fillna
dataset['Outlet_Size']=dataset['Outlet_Size'].fillna("Medium")
```

```python
# check the duplicate values available in the data
dataset.duplicated().sum()
```

```python
#Convert Outlet_Establishment_Year to Outlet_Age
current_year = 2025
dataset['Outlet_Age']=current_year - dataset['Outlet_Establishment_Year']
```

```python
#value count of Item_Type	
dataset['Item_Type'].value_counts()
```

```python
# check the unique values available in the Item_Fat_Content column
dataset['Item_Fat_Content'].unique()
```

```python
# replace LF to Low Fat
dataset['Item_Fat_Content'].replace("LF","Low Fat", inplace=True)
```

```python
# replace low fat to Low Fat
dataset['Item_Fat_Content'].replace("low fat","Low Fat", inplace=True)
```

```python
# replace reg to Regular
dataset['Item_Fat_Content'].replace("reg","Regular", inplace=True)
```

```python
# check the value from 45 to 50
dataset.loc[45:50,['Item_Fat_Content']]
```

```python
# check the unique values available in the Outlet_Size column
dataset['Outlet_Size'].unique()
```

```python
# check the unique values available in the Outlet_Type column
dataset['Outlet_Type'].unique()
```

```python
# check the unique values available in the Item_Type column
dataset['Item_Type'].unique()
```

```python
# round of the Item_MRP by 2 digits
dataset['Item_MRP'] = np.round(dataset['Item_MRP'],2)
```

```python
# round of the Item_Outlet_Sales by 2 digits
dataset['Item_Outlet_Sales'] = np.round(dataset['Item_Outlet_Sales'],2)
```

```python
# round of the Item_Visibility by 2 digits
dataset['Item_Visibility'] = np.round(dataset['Item_Visibility'],2)
```

```python
conn = msc.connect( host = "localhost",
             user = "root",
             password = "New@95697",
             port = 3306,
             database = "Blinkt_Proj"
                  )
cur = conn.cursor()
```

```python
query = """
CREATE TABLE Sales_Data(
    Column1 VARCHAR(50),
    Item_Weight FLOAT,
    Item_Fat_Content VARCHAR(20),
    Item_Visibility FLOAT,
    Item_Type VARCHAR(50),
    Item_MRP FLOAT,
    Outlet_Identifier VARCHAR(20),
    Outlet_Establishment_Year INT,
    Outlet_Size VARCHAR(20),
    Outlet_Location_Type VARCHAR(20),
    Outlet_Type VARCHAR(30),
    Item_Outlet_Sales FLOAT,
    Outlet_Age INT
);
"""

cur.execute(query)
conn.commit()


#conn.close()
```

```python
# Verify table created
cur.execute("SHOW TABLES")
for table in cur:
    print(table)
```

```python
# ✅ Insert data into table
insert_query = """
INSERT INTO Sales_Data (
    Column1,
    Item_Weight,
    Item_Fat_Content,
    Item_Visibility,
    Item_Type,
    Item_MRP,
    Outlet_Identifier,
    Outlet_Establishment_Year,
    Outlet_Size,
    Outlet_Location_Type,
    Outlet_Type,
    Item_Outlet_Sales,
    Outlet_Age
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)    # %s = slot for values
"""

for row in dataset.itertuples(index=False):
    cur.execute(insert_query, tuple(row))

conn.commit()

print("✅ Data uploaded successfully!")

cur.close()
conn.close()
```
